```python
#Formação Cientista de Dados & Engenharia de Dados
from pyspark.sql import SparkSession
```

```python
carros = spark.read.csv("/FileStore/tables/Carros-5.csv", header=True, sep=";",inferSchema=True)
carros.show(5)
```
